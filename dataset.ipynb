```python
import cv2 as cv
import utils.dataset as dataset
```

```python
mydataset = dataset.MyDataset('./data/VOC2007/VOCdevkit/VOC2007/JPEGImages', './data/VOC2007/VOCdevkit/VOC2007/Annotations')

mydataset.object_ids
```

```python
example = iter(mydataset)
```

```python
ret = next(example)
ret = next(example)
ret = next(example)
```

```python
import matplotlib.pyplot as plt

plt.subplot(1, 2, 1)
plt.imshow(ret['image'])
plt.subplot(1, 2, 2)
plt.imshow(ret['origin_image'])
```

```python
bboxes = ret['label']

for bbox in bboxes:
    class_id = bbox.get('class')
    x1 = int(bbox.get('x1'))
    y1 = int(bbox.get('y1'))
    x2 = int(bbox.get('x2'))
    y2 = int(bbox.get('y2'))
    class_name = mydataset.object_names[class_id]

    cv.rectangle(ret['image'], (x1, y1), (x2, y2), (0, 255, 0), 2)
    cv.putText(ret['image'], class_name, (x1, y1),
               cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1)

plt.imshow(ret['image'])
```
